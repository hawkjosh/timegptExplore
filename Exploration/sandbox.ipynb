```python
import os

from TokenUsage import showTokenUsage
from dotenv import load_dotenv
from GenerateData import generate_basic_data, generate_data_with_ex_vars
from nixtlats import TimeGPT

load_dotenv()
```

```python
timegpt = TimeGPT(token=os.getenv("TIMEGPT_TOKEN"))
timegpt.validate_token()
```

```python
showTokenUsage()
```

```python
start = "2023-12-06"
end = "2024-03-04"
low = 75
high = 125

df_basic = generate_basic_data(start, end, low, high)
df_basic
```

```python
df_basic_anomalies = timegpt.detect_anomalies(
    df_basic,
    time_col="timestamp",
    target_col="value",
)
df_basic_anomalies
```
